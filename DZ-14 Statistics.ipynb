```python
import pandas as pd
import numpy as np
df = pd.read_csv('horse_data.csv', na_values=('?'))
df.columns = (['surgery?',                                      # 'Хирургия? (1 -да/2 -нет)',
                       'Age',                                   # 'Возраст (1-взрослая/2-молодая)', 
                        'Hospital_Number',                      # 'Номер госпитализации (id)',
                       'rectal_temperature',                    # 'Ректальная температура (норм 37,8)',   
                        'pulse',                                # 'Пульс (норм 30-40)',
                       'respiratory_rate',                      # 'Частота дыхания (норм 8-10)', 
                        'temperature_of_extr',                  # 'температура конечностей (1-норм, 2-горячо, 3-прохладно, 4-холодно)',
                       'peripheral_pulse',                      # 'периферический пульс (1-норм/4-отсутствие)'
                        'mucous_membranes',                     # 'слизистые оболочки (1-норм/6-темно-синий)',
                       'capillary_refill_time',                 # 'время пополнения капилляров (1 -до 3сек/2 -больше 3 сек)'
                        'pain',                                 # 'боль(1 -норм/5 -непрерывная сильная боль)',
                       'peristalsis',                           # 'перистальтика(1-гипермотиль, 2-норм, 3-гипо, 4-отсутствует)'
                        'abdominal_distension',                 # 'вздутие живота(1-нет/4-сильно)',
                       'nasogastric_tube',                      # 'назогастральный зонд(1 -нет/2 -существенно)'
                        'nasogastric_reflux',                   # 'назогастральный рефлюкс(1-нет, 2-больше 1 л, 3-меньше 1 л)',
                       'nasogastric_reflux_PH',                 # 'назогастральный рефлюкс РН(3-4 - норм)'
                        'rectal_examination',                   # 'ректальное исследование(1 -норм/4 -отсутствует)',
                       'abdomen',                               # 'брюшная полость(1-норм, 2-другое, 3--5-проблемы)'
                        'packed_cell_volume',                   # 'объем упакованных клеток(30-50-норм)',
                       'total_protein',                         # 'общий белок(6-7,5-норм)'
                        'abdominocentesis_appearance',          # 'внешний вид абдоминоцентеза(1-норм, 2 и 3-проблемы)',
                       'abdomcentesis_total_protein',           # 'абдоминоцентез общий белок(чем выше, тем хуже)'
                        'outcome',                              # 'гибель (1-жив, 2-мертв, 3-усыплен)',
                       'surgical_lesion?',                      # 'хирургическое повреждение?(1 -да/2 -нет)'
                        'type_of_lesion',                       # 'тип повреждения(1-11-орган поражения/1-4-тип_смерти/1,2,0/0-10)', 
                       '25',                                    #'25',
                        '26',                                   # '26',
                       'cp_data']) 
df_analyse = df[['surgery?', 'Age', 'pulse', 'respiratory_rate', 'pain', 'packed_cell_volume', 'outcome', 'surgical_lesion?']]
df_analyse.describe()
```

```python
df_analyse.sort_values('surgery?')
```

```python
# df_analyse.info()
df_analyse = df_analyse.dropna(thresh=6)
```

```python
df_analyse.info()
```

```python
df_analyse.pulse.fillna(df_analyse.pulse.median(), inplace=True)
df_analyse.respiratory_rate.fillna(df_analyse.respiratory_rate.median(), inplace=True)
df_analyse.pain.fillna(df_analyse.pain.median(), inplace=True)
df_analyse.packed_cell_volume.fillna(df_analyse.packed_cell_volume.median(), inplace=True)
```

```python
df_analyse.info()
```

```python
df_analyse.dropna().info()
```

```python
q1 = df.pulse.quantile(0.25)
q3 = df.pulse.quantile(0.75)
iqr = q3 - q1
lower_bound = q1 - (1.5 * iqr) 
upper_bound = q3 + (1.5 * iqr)
remove_outliers = df_analyse[df_analyse.pulse.between(lower_bound, upper_bound, inclusive=True)].sort_values('pulse')
pd.concat([df_analyse, remove_outliers]).drop_duplicates(keep=False)
```
